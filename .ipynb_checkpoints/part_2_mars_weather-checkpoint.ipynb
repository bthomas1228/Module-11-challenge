```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
rows = soup.find_all('tr', class_='data-row')
rows
#pretty print rows to check extraction
#for row in rows:
    #print(row.text)
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
data = []

# Loop through the scraped data to create a list of rows

for row in rows:
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text)
    data.append(row_data)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data)
table_df = df.rename(columns={0:'id', 1:'terrestrial_date', 2:'sol', 3:'ls', 4:'month', 5:'min_temp', 6:'pressure'})
```

```python
# Confirm DataFrame was created successfully
table_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
table_df.dtypes
```

```python
# Change data types for data analysis
table_df['sol'] = table_df['sol'].astype('int')
table_df['ls'] = table_df['ls'].astype('int')
table_df['month'] = table_df['month'].astype('int')
table_df['min_temp'] = table_df['min_temp'].astype('float')
table_df['pressure'] = table_df['pressure'].astype('float')
table_df['terrestrial_date'] = pd.to_datetime(table_df['terrestrial_date'])
```

```python
# Confirm type changes were successful by examining data types again
table_df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
num_months = len(table_df['month'].unique())
num_months
```

```python
# 2. How many sols (Martian days) worth of data are there?
table_df['sol'].count()
```

```python
# 3. What is the average minimum temperature by month?
min_temp = table_df.groupby('month')['min_temp'].mean()
min_temp
```

```python
# Plot the average minimum temperature by month
min_temp.plot(kind="bar", figsize=(7,5), ylabel="Temperature in Celsius", legend=False)
plt.title("Average Minimum Temperature by Month")
plt.show()
plt.tight_layout()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
min_temp_ascending = min_temp.sort_values(ascending=True)
min_temp_ascending.plot(kind="bar", figsize=(7,5), ylabel="Temperature in Celsius", legend=False)
plt.title("Average Minimum Temperature by Month")
plt.show()
plt.tight_layout()

#The coldest month is "3" and the hottest month is "8".
```

```python
# 4. What is the average pressure by month?
pressure_ave = table_df.groupby('month')['pressure'].mean()
pressure_ave
```

```python
# Plot the average pressure by month
pressure_ave.plot(kind="bar", figsize=(7,5), ylabel="Atmospheric Pressure", legend=False)
plt.title("Average Pressure by Month")
plt.show()
plt.tight_layout()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
pressure_ascending = pressure_ave.sort_values(ascending=True)
pressure_ascending.plot(kind="bar", figsize=(7,5), ylabel="Atmospheric Pressure", legend=False)
plt.title("Average Pressure by Month")
plt.show()
plt.tight_layout()

#The month with the lowest atmospheric pressure is '6' and the month with the highest atmospheric pressure is '9'.
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
earth_days = table_df['terrestrial_date']
temperature = table_df['min_temp']
plt.plot(earth_days, temperature)
plt.title("Minimum Temperature over Time")
plt.xlabel("Number of Terrestrial Days")
plt.ylabel("Minimum Temperature")
plt.show()
plt.tight_layout()

#### Minimum Temperature
```

```python
earth_days = table_df['terrestrial_date']
solar_longitude = table_df['ls']
plt.plot(earth_days, solar_longitude)
plt.title("Solar Longitude over Time")
plt.xlabel("Number of Terrestrial Days")
plt.ylabel("Solar Longitude")
plt.show()
plt.tight_layout()
```

#### Atmospheric Pressure

The month with the lowest atmospheric pressure is '6' and the month with the highest atmospheric pressure is '9'.
'9' also has one of the higher temperatures whereas '6' has an average temperature in the middle of the range. 
This possibly suggests that high atmospheric pressure predicts high temperatures but it doesn't hold for values other than at the very high range. Pressure is also likely cyclical but the data suggests it doesn't follow the same cycle (high, low) as temperature. 



#### Year Length

If we assume the Mars data in the table spans the equivalent of 1 Martian year, then the number of earth days in 1 year is 2021 (calculated by subtracting the first date of data from the last date in the cell directly following the plot of minimum temperature over terrestrial days).

However, in looking at the Mars temperature data in the above graph, it displays a cyclical pattern suggestive of full years, with each year represented by a full peak and valley. In that case, visually you could estimate a Martian year to be the time between two peaks or two valleys (ie ~2 earth years = 1 Martian year). This would assume that the monthly data (e.g. average minimum temperature per month) was an aggreggation of data for the same month over several years instead of just one year.

To further test this hypothesis, I polotted Solar Longitude (ls) over terrestrial days and it also showed a similar cyclical pattern to the minimum temperature pattern. Estimating the time between valleys, the total terrestrial days in one Martian year = ~2 yrs. To cacluate this using the data, you could extract the rows representing the lowest ls and subtract the two nearest dates to calculate a year.
